# bitwise AND (&)

```python
a = 10
b = 5
bin (a & b)
```

```python
a = 10
b = 5
bin (a | b)
```

```python
a = 10
b = 5
a | b
```

# bitwise NOT (~)
- returns one's compliment of a number
- one's compliment means each bit will be complimented in the given number and compliment means the bit will be flipped, meaning 0 becomes 1 andd 1 becomes 0.

```python
a = 10
bin(10)
```

```python
~a 
```

```python
a = 10
b = 9
a ^ b
```

```python
# break - break out of the loop
numbers = range(0,21)
even_list = []
odd_list = []
for i in numbers:
    if i == 14:
         break
    if i%2==0:
        even_list.append(i)
    else:
        odd_list.append(i)
print (f"the even list is {even_list}")
print (f"the odd list is {odd_list}")
```

```python
# continue - skip at a particular condition alone
numbers = range(0,21)
even_list = []
odd_list = []
for i in numbers:
    if i == 14:
         continue
    if i%2==0:
        even_list.append(i)
    else:
        odd_list.append(i)
print (f"the even list is {even_list}")
print (f"the odd list is {odd_list}")
```

# function

```python
def agegroup():
    age = int(input("please enter your age to know your age group"))
    if age <=1:
        print(f"the entered age is {age}, therefore, you are an infant")
    elif (age >=2 and age <=4):
        print(f"the entered age is {age}, therefore, you are a toddler")
    elif (age >=5 and age <=12):
        print(f"the entered age is {age}, therefore, you are a child")
    elif (age >=13 and age <=19):
        print(f"the entered age is {age}, therefore, you are a teenager")
    elif (age >=20 and age <=39):
        print(f"the entered age is {age}, therefore, you are a adult")
    elif (age >=40 and age <=59):
        print(f"the entered age is {age}, therefore, you are a middle age")
    else:
        print(f"the entered age is {age}, therefore, you are a senior age")
```

```python
t= [1,3,4,8,9]
square = lambda x: x**2
out_put= list(map(square,t))
out_put
```
